```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
# import cv2
from scipy.stats import stats
import matplotlib.image as mpimg
```

Kompresja obrazu uzyskana za pomocą algorytmu PCA z użyciem 16 współrzędnych.
Wyjściowy obraz o rozdzielczości 285 × 285 (po lewej stronie) został najpierw podzielony na kwadraty 8 × 8, a następnie zareprezentowany za pomocą wektorów o wymiarze 8 · 8 = 64 (analogicznie jak w formacie JPG). PCA został użyty do opisu takiego zbioru danych za pomocą 16-wymiarowej podprzestrzeni. Obraz po prawej przedstawia kompresję uzyskaną przez wykorzystanie rekonstrukcji
z tej podprzestrzeni.

```python
import numpy as np
from skimage import io
img  = io.imread('boat.jpg')
```

Rozłożenie obrazu na kanały RGB

```python
#Splitting into channels
blue = img[:,:,0]
green = img[:,:,1]
red = img[:,:,2] 
# Plotting the images
fig = plt.figure(figsize = (15, 7.2)) 
fig.add_subplot(131)
plt.title("Blue Channel")
plt.imshow(blue)
fig.add_subplot(132)
plt.title("Green Channel")
plt.imshow(green)
fig.add_subplot(133)
plt.title("Red Channel")
plt.imshow(red)
plt.show()
```

Przeskalowanie kolorów do zakresu [0,1]

```python
df_blue = blue/255
df_green = green/255
df_red = red/255
```

Wykonuje alorytm PCA z n_components=16.

Aby zreprodukować drugi obraz usta n_components=50.

```python
pca_b = PCA(n_components=16)
pca_b.fit(df_blue)
trans_pca_b = pca_b.transform(df_blue)

pca_g = PCA(n_components=16)
pca_g.fit(df_green)
trans_pca_g = pca_g.transform(df_green)

pca_r = PCA(n_components=16)
pca_r.fit(df_red)
trans_pca_r = pca_r.transform(df_red)
```

```python
print(trans_pca_b.shape)
print(trans_pca_r.shape)
print(trans_pca_g.shape)
```

```python
print(f"Blue Channel : {sum(pca_b.explained_variance_ratio_)}")
print(f"Green Channel: {sum(pca_g.explained_variance_ratio_)}")
print(f"Red Channel  : {sum(pca_r.explained_variance_ratio_)}")
```

Histogram reprezentujący ważność komponentów.

```python
fig = plt.figure(figsize = (15, 7.2)) 
fig.add_subplot(131)
plt.title("Blue Channel")
plt.ylabel('Variation explained')
plt.xlabel('Eigen Value')
plt.bar(list(range(1,17)),pca_b.explained_variance_ratio_)
fig.add_subplot(132)
plt.title("Green Channel")
plt.ylabel('Variation explained')
plt.xlabel('Eigen Value')
plt.bar(list(range(1,17)),pca_g.explained_variance_ratio_)
fig.add_subplot(133)
plt.title("Red Channel")
plt.ylabel('Variation explained')
plt.xlabel('Eigen Value')
plt.bar(list(range(1,17)),pca_r.explained_variance_ratio_)
plt.show()
```

```python
b_arr = pca_b.inverse_transform(trans_pca_b)
g_arr = pca_g.inverse_transform(trans_pca_g)
r_arr = pca_r.inverse_transform(trans_pca_r)
print(b_arr.shape, g_arr.shape, r_arr.shape)
```

```python
img_reduced = np.zeros(img.shape)
img_reduced[:,:,0] = b_arr
img_reduced[:,:,1] = g_arr
img_reduced[:,:,2] = r_arr
img_reduced[img_reduced<0] = 0
img_reduced[img_reduced>1] = 1
print(img_reduced.shape)
```

Wizualizacja obrazu po kompresji.

```python
plt.imshow(img_reduced, cmap=plt.cm.viridis)
plt.axis('off')
plt.title("Base image")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.title('Compressed', fontsize=20)
plt.xticks([])
plt.yticks([])
plt.imshow(img_reduced)
plt.show()

plt.figure(figsize=(8, 6))
plt.title('Original', fontsize=20)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```
