Kompresja obrazu uzyskana za pomocą algorytmu k-means z 16 grupami. Klastrowanie zostało wykonane na 3-wymiarowym zbiorze kolorów obrazu.

Wynikowy obraz zawiera tylko 16 kolorów (każdy kolor został zastąpiony przez kolor reprezentujący odpowiadający mu klaster).

```python
import numpy as np
from skimage import io
img  = io.imread('boat.jpg')
img_data = (img / 255.).reshape(-1, 3)
img_data.shape
```

```python
plt.imshow(img)
plt.grid(b=None)
plt.axis('off')
plt.show()
```

```python
from sklearn.cluster import KMeans
import time

start = time.time()
kmeans = KMeans(n_clusters=16, max_iter=300)
kmeans.fit(img_data)
closest = kmeans.predict(img_data)
k_colors = kmeans.cluster_centers_[closest]
print(f'elapsed: {time.time() - start}')

```

```python
k_img = np.reshape(k_colors,(img.shape))


plt.figure(figsize=(8, 6))
plt.title('Compressed', fontsize=20)
plt.xticks([])
plt.yticks([])
plt.imshow(k_img)
plt.show()

plt.figure(figsize=(8, 6))
plt.title('Original', fontsize=20)
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.show()
```
