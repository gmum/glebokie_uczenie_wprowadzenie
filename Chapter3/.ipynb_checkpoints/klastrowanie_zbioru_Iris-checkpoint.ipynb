```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
```

Wyniki klastrowania zbioru Iris (reprezentowanego za pomocą 2 pierwszych współrzędnych) z użyciem k-means z parametrem k równym 1, 2, 3, 4, 5, odpowiednio. Dla porównania na pierwszym rysunku przedstawiono podział referencyjny zbioru. Poniżej przedstawiono wykres funkcji koszu k-means oraz miarę zgodności uzyskanego podziału z podziałem referencyjnym w zależności od
liczby grup.

```python
from sklearn import datasets
iris = datasets.load_iris()

X = iris.data[:,0:2]  
y = iris.target

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Dwie pierwsze współrzędne ze zbioru danych Iris \n Prawdziwy przydział do klastrów", fontsize=15)
plt.show()
```

```python
iris.feature_names
```

```python
adjusted_rand_score_array = [None]
SSE_array = [None]
for i in range(1,11):
    km = KMeans(n_clusters=i, random_state=0)
    labels_km = km.fit_predict(X)
    acc = metrics.adjusted_rand_score(labels_km, y)
    adjusted_rand_score_array.append(acc)
    SSE_array.append(km.inertia_)
    
    
    
    # Step size of the mesh. Decrease to increase the quality of the VQ.
    h = .02     # point in the mesh [x_min, x_max]x[y_min, y_max].

    # Plot the decision boundary. For that, we will assign a color to each
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Obtain labels for each point in mesh. Use last trained model.
    Z = km.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(8, 6))
    plt.imshow(Z, interpolation="nearest",
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap=plt.cm.Paired, aspect="auto", origin="lower")

    centroids = km.cluster_centers_
    plt.scatter(centroids[:, 0], centroids[:, 1], marker="x", s=169, linewidths=3,
                color="w", zorder=10)
    

    plt.scatter(X[:, 0], X[:, 1], c=labels_km)
    plt.title("Dwie pierwsze współrzędne ze zbioru danych Iris \n Wynik algorytmu k-means z k = " +str(i), fontsize=15)
    plt.show()
```

```python
t=range(0,11)
plt.figure(figsize=(10,6));
plt.plot( adjusted_rand_score_array, "o-", label="Rand score")
# plt.plot( adjusted_mutual_info_score_array, "o-", label="adjusted_mutual_info")
plt.xlim(0,11)
plt.xticks(range(len(t)), t, size='small')
# plt.legend()
plt.title("Porównanie miary Rand score w zależności od ilości klastrów", fontsize=15)
plt.ylabel('Rand score', fontsize=12)
plt.xlabel('Ilość klastrów', fontsize=12)

plt.title("Rand score", fontsize=15)
plt.show()
```

```python
t=range(0,11)
plt.figure(figsize=(10,6));
plt.plot( SSE_array, "o-", label="Rand score")
plt.xlim(0,11)
plt.xticks(range(len(t)), t, size='small')
# plt.legend()
plt.title("Porównanie miary sum of squared w zależności od ilości klastrów", fontsize=15)
plt.ylabel('Sum of squared', fontsize=12)
plt.xlabel('Ilość klastrów', fontsize=12)
plt.show()
```
