```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Algorytm PCA wyznacza kierunki największej wariancji danych. Wykorzystując
te kierunki dokonujemy obrotu układu współrzędnych w ten sposób, że kolejne współrzędne nowego układu stanowią kierunki wyznaczone przez PCA w kolejności ważności. Redukcja wymiarowości polega na zrzutowaniu danych do pierwszych k współrzędnych.

```python
n = 1000
mean = [2, 4]
A=np.array([[1,1],[0,2]])
cov = np.matmul(A, np.transpose(A)) #cov = A*A^T
X = np.random.multivariate_normal(mean, cov, n)


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.scatter(X[:, 0], X[:, 1])
plt.axis('equal')
plt.xlim=(0,10)
plt.ylim=(0,10)
plt.show()
```

Wyznaczamy komponenty główne i ich istotność

```python
cov_mat = np.cov(X.T)
print('Macierz kowariancji: \n%s' %cov_mat)

e_values, e_vectors = np.linalg.eigh(cov_mat)

print('Eigenvectors \n%s' %e_vectors)
print('\nEigenvalues \n%s' %e_values)

m = np.mean(X,0)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.scatter(X[:, 0], X[:, 1])
soa =np.array( [ [m[0] + 0, m[1] + 0, 2*np.sqrt(e_values[1])*e_vectors[0,1], 2*np.sqrt(e_values[1])*e_vectors[1,1]],
                 [m[0] + 0, m[1] + 0, 2*np.sqrt(e_values[0])*e_vectors[0,0], 2*np.sqrt(e_values[0])*e_vectors[1,0]]] )

X1,Y1,U1,V1 = zip(*soa)
ax.quiver(X1,Y1,U1,V1,angles='xy',scale_units='xy',scale=1)
plt.axis('equal')
plt.show()
```

Wizualizacja danych po transformacji.

```python
from sklearn.decomposition import PCA
model = PCA()

pca_features = model.fit_transform(X) 

xs = pca_features[:,0]
ys = pca_features[:,1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
plt.scatter(xs, ys)
plt.axis('equal')
plt.show()
```
