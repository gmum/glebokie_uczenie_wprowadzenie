```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from scipy import optimize
```

Estymacja rozkładu danych pochodzących ze standardowego rozkładu normalnego opisanego histogramem. Metoda największej wiarygodności daje optymalną estymację.

```python
N = 100000
t = np.arange(-6, 6, 0.05)
n=stats.norm()
data = n.rvs(size=N)
```

Wizualizacja gęstości rozkładu normalnego z różnymi parametrami.

```python
mu= np.mean(data)
sigma = np.std(data)

mu_values = [0, 0.3, 0.8, -2]
sigma_values = [1.9, 1.5, 1., 1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 5))
ax.hist(data, bins=25, density=True, histtype='stepfilled', alpha=0.5, label='histogram')
ax.legend(loc='best', frameon=False)
for mu, sigma in zip(mu_values, sigma_values):
    dist = stats.norm(mu, sigma)
    ax.plot(t, dist.pdf(t), ls="--", label=r'$\mu=%.1f,\ \sigma=%.1f$' % (mu, sigma))
ax.legend()
plt.show()
```

Wizualizacja estymatorów największej wiarygodności.

```python
mu= np.mean(data)
sigma = np.std(data)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 5))
ax.hist(data, bins=25, density=True, histtype='stepfilled', alpha=0.5, label='histogram')
ax.legend(loc='best', frameon=False)
ax.plot(t, stats.norm.pdf(t,mu, sigma), 'k-', lw=2, label=r'$\mu=%.1f,\ \sigma=%.1f$' % (mu, sigma))
ax.legend()
plt.show()
```
