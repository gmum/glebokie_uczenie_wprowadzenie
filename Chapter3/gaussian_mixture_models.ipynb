```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import numpy as np
from scipy import linalg
import itertools
```

Estymacja rozkładu danych za pomocą GMM z różną liczbą komponentów.

```python
from sklearn.datasets import make_moons
X, y_true = make_moons(200, noise=.05, random_state=0)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))
ax.scatter(X[:, 0], X[:, 1], s=40, cmap='viridis')
ax.axis('equal')
plt.show()
```

```python
from sklearn import mixture
gmm = mixture.GaussianMixture(n_components=2).fit(X)
labels = gmm.predict(X)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))
ax.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis')
ax.axis('equal')
plt.show()
```

```python
color_iter = itertools.cycle(['navy', 'turquoise', 'cornflowerblue', 'darkorange'])
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))
for i, (mean, cov, color) in enumerate(zip(gmm.means_, gmm.covariances_, color_iter)):
    v, w = linalg.eigh(cov)
    plt.scatter(X[labels == i, 0], X[labels == i, 1], 20, color=color)

    # Plot an ellipse to show the Gaussian component
    angle = np.arctan2(w[0][1], w[0][0])
    angle = 180. * angle / np.pi  # convert to degrees
    v = 2. * np.sqrt(2.) * np.sqrt(v)
    ell = mpl.patches.Ellipse(mean, v[0], v[1], 180. + angle, color=color)
    ell.set_alpha(.5)
    ax.add_artist(ell)
    
ax.axis('equal')
plt.show()
    
```
