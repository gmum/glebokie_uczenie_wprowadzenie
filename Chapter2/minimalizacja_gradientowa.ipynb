## Minimalizacja gradientowa (gradient descent GD).

Ogólny schemat minimalizacji gradientowej można przedstawić w kilku krokach:

*  wybierz (losowo) punkt startowy $x_0 \in \mathbb{R}^D$, $i=0$
*  ustal wielkość kroku (learning rate) $h>0$ i warunek stopu -- domyślnie jest to ilość planowanych kroków $k$,
* wykonaj $k$ krotnie:  
    $$
    x_{i+1} =x_i-h \nabla f(x_i), \, i=i+1,
    $$
* zwróć $x_k$.

W ten sposób otrzymujemy zarówno minimum lokalne funkcji $f$ oraz punkt, który je realizuje.


## wielkość kroku $h$
Istotnym parametrem minimalizacji gradientowej jest wielkość kroku $h$. Jeżeli jest za mały to zbieżność może być zbyt wolna, bądź wartość znalezionego minimum lokalnego może być dalekie od wartości minimum globalnego. Jeżeli jest za duży, to minimalizacja gradientowa może nie być zbieżna -- funkcja którą chcemy zminimalizować\footnote{Zwyczajowo mówi się w informatyce o funkcji kosztu.} nie będzie maleć. 

Aby to zaobserwować, rozpatrzmy 
$$
f(x)=ax^2, 
$$

gdzie $a>0$ jest ustalonym parametrem.

Napisz (od podstaw) algorytm znajdowania minimum metodą gradientową.
Zaprezentuj jego działanie na przykładzie funkcji $f(x)=x^2$.

```python
f = lambda x: x**2

x = np.linspace(-5, 5, 200)
y = f(x)


plt.figure(figsize=(7, 7))
plt.plot(x, y, '--k', ms=10, label = r"$f(x)=x^2$");
plt.title("Histogram", fontsize=15)
plt.legend(loc="lower right", fontsize=12)
plt.show()


plt.show()
```

```python
def step_gradient_1d(x_current, learningRate):
    x_gradient = 2*x_current
    new_x = x_current - (learningRate * x_gradient)
    
    plt.arrow(x_current, f(x_current), - (learningRate * x_gradient), -(f(x_current)-f(new_x)),
              head_width=0.04, head_length=0.04, ec="red", color="red", length_includes_head=True)
#     plt.scatter(x_current, f(x_current), c = "blue")
        
    return new_x
def gradient_descent_runner_1d(starting_x, learning_rate, num_iterations):
    x = starting_x
    print(x)
    for i in range(num_iterations):
        x = step_gradient_1d(x,learning_rate)
    return x
```

```python
x = np.linspace(-1, 1, 200)
y = f(x)

plt.figure(figsize=(7, 5))
plt.plot(x, y, '--k', ms=10, label = r"$f(x)=x^2$");

learning_rate = 0.1
initial_x = 0.8
num_iterations = 10
x = gradient_descent_runner_1d(initial_x, learning_rate, num_iterations)
plt.text(0.0, 0.8, r"$h<\frac{1}{a}$", fontsize=15)
plt.text(0.0, 0.7, r"$v_{i+1} =v_i-h \nabla f(v_i)$", fontsize=15)
plt.legend(loc="lower right", fontsize=12)
plt.scatter([0.8], [0.8**2], c="k");
plt.text(0.8, 0.55, r"$v_{0}$", fontsize=15)


plt.title("Minimalizacja gradientowa", fontsize=15)
plt.show()
```

Porównanie minimalizacji gradientowej i iteracyjnej na przykładzie Przykład 2.1.(z książki).

```python
import numpy as np
import matplotlib.pyplot as plt

chi2 = lambda x,y: np.abs( x - y ) + np.abs( x - 1 ) + np.abs(y)

x = np.arange(-12,12,0.02)
y = np.arange(-12,12,0.02)

X,Y= np.meshgrid(x,y)

Z = chi2(X,Y)

# plt.figure()

plt.figure(figsize=(7, 7))

CS = plt.contour(X,Y,Z, levels=12)

# plt.plot([5],[5],"o")

#####################################

def step_gradient_2d(x_current, y_current, learningRate):
    x_gradient = np.sign(x_current-y_current) + np.sign(x_current - 1)
    y_gradient = -np.sign(x_current-y_current) + np.sign(y_current)
   
    new_x = x_current - (learningRate * x_gradient)
    new_y = y_current - (learningRate * y_gradient)
   
    plt.arrow(x_current, y_current, - (learningRate * x_gradient), - (learningRate * y_gradient), 
              head_width=0.5, head_length=0.5, ec="red", color="red", length_includes_head=True)
   
    return [new_x, new_y]
def gradient_descent_runner_2d(starting_x, starting_y, learning_rate, num_iterations):
    x = starting_x
    y = starting_y
    for i in range(num_iterations):
        x, y = step_gradient_2d(x, y, learning_rate)
        #print(x, y)
    return [x, y]


learning_rate = 0.4
initial_x = -10 # initial y-intercept guess
initial_y = 5 # initial slope guess
num_iterations = 12
[x, y] = gradient_descent_runner_2d(initial_x, initial_y, learning_rate, num_iterations)

plt.scatter([initial_x], [initial_y], c="k");
# plt.text(1, 0.6, r"$v_{0}$", fontsize=15)

#####################################
plt.axis('equal')
plt.title("Minimalizacja gradientowa", fontsize=15)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

chi2 = lambda x,y: np.abs( x - y ) + np.abs( x - 1 ) + np.abs(y)

x = np.arange(-12,12,0.01)
y = np.arange(-12,12,0.01)

X,Y= np.meshgrid(x,y)

Z = chi2(X,Y)

# plt.figure()

plt.figure(figsize=(7, 7))

CS = plt.contour(X,Y,Z, levels=12)

# plt.plot([5],[5],"o")

#####################################

def step_gradient_2d(x_current, y_current, learningRate, it):
    x_gradient = np.sign(x_current-y_current) + np.sign(x_current - 1)
    y_gradient = -np.sign(x_current-y_current) + np.sign(y_current)
   
    if it % 2 == 0:
        new_x = x_current 
        new_y = (x_current+1)/2
    else:
        new_x = y_current/2
        new_y = y_current 
        
    plt.arrow(x_current, y_current, new_x- x_current, new_y - y_current, 
              head_width=0.5, head_length=0.5, ec="red", color="red", length_includes_head=True)
   
    return [new_x, new_y]
def gradient_descent_runner_2d(starting_x, starting_y, learning_rate, num_iterations):
    x = starting_x
    y = starting_y
    for i in range(num_iterations):
        x, y = step_gradient_2d(x, y, learning_rate, i)
        print(i)
    return [x, y]


learning_rate = 0.7
initial_x = -10 # initial y-intercept guess
initial_y = 5 # initial slope guess
num_iterations = 6

[x, y] = gradient_descent_runner_2d(initial_x, initial_y, learning_rate, num_iterations)

plt.scatter([initial_x], [initial_y], c="k");
# plt.text(1, 0.6, r"$v_{0}$", fontsize=15)

#####################################
plt.axis('equal')
plt.title("Minimalizacja iteracyjna", fontsize=15)
plt.show()
```
