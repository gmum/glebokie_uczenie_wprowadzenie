```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs
```

Wizualizacja bariery decyzyjnej logistycznej regresji i SVM z domy≈õlnymi parametrami.

```python
centers = [(0, 0), (5, 5)]
X, y = make_blobs(n_samples=200, centers=centers, shuffle=False, random_state=42, cluster_std=0.5)
X = np.vstack((X, [2,2]))
y = np.concatenate((y, [1]))

X = np.vstack((X, [1.5,1.5]))
y = np.concatenate((y, [1]))

plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Dwie klasy w tym jedna zawierajƒÖca element odstajƒÖcy", fontsize=15)
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
```

# Perceptron

```python
from sklearn.linear_model import Perceptron


per_clf = Perceptron(random_state=0)
per_clf.fit(X,y)

plt.figure(figsize=(8, 5))
plot_decision_regions(X, y, per_clf)
plt.title("Klasyfikacyjna metodƒÖ perceptrona", fontsize=15)
plt.show()
```

# Logistic Regresion

```python
from sklearn.linear_model import LogisticRegression

lr_clf = LogisticRegression(random_state=0)
lr_clf.fit(X,y)

plt.figure(figsize=(8, 5))
plot_decision_regions(X, y, lr_clf)
plt.title("Klasyfikacyjna metodƒÖ Logistycznej Regresji", fontsize=15)
plt.show()
```

# Linear SVM

```python
from sklearn.svm import SVC

svm_clf = SVC(kernel='linear', random_state=0, C=10)
svm_clf.fit(X,y)

plt.figure(figsize=(8, 5))
plot_decision_regions(X, y, svm_clf)
plt.title("Klasyfikacyjna metodƒÖ liniowego SVM", fontsize=15)
plt.show()
```
