```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from scipy import stats

from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import  metrics
```

```python
X = np.array([i for i in range(10,30)])
y = np.concatenate([np.ones(10), np.zeros(10)])

X = np.append(X, np.array([i for i in range(60,70)]))
y = np.append(y, np.zeros(10))

fig = plt.figure(figsize=(10,5))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")
plt.show()
```

Uczymy modele regresji z rożnymi parametrami.

```python
X = X.reshape(-1,1)
model1 = linear_model.LinearRegression()
model1.fit(X, y)

model2 = make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression())
model2.fit(X, y)

model3 = make_pipeline(PolynomialFeatures(3), linear_model.LinearRegression())
model3.fit(X, y)

model4 = make_pipeline(PolynomialFeatures(4), linear_model.LinearRegression())
model4.fit(X, y)

model5 = make_pipeline(PolynomialFeatures(5), linear_model.LinearRegression())
model5.fit(X, y)

model6 = make_pipeline(PolynomialFeatures(6), linear_model.LinearRegression())
model6.fit(X, y)

model7 = make_pipeline(PolynomialFeatures(7), linear_model.LinearRegression())
model7.fit(X, y)

model8 = make_pipeline(PolynomialFeatures(8), linear_model.LinearRegression())
model8.fit(X, y)

model9 = make_pipeline(PolynomialFeatures(9), linear_model.LinearRegression())
model9.fit(X, y)

model10 = make_pipeline(PolynomialFeatures(10), linear_model.LinearRegression())
model10.fit(X, y)
```

Wizualizacja wyników

```python
x_f_lr = np.vstack(np.linspace(10, 70, 2000))

fig = plt.figure(figsize=(10,5))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")

plt.plot(x_f_lr, model1.predict(x_f_lr), label='Model liniowy')
# plt.plot(x_f_lr, model2.predict(x_f_lr), label='polinominal 2')
# plt.plot(x_f_lr, model3.predict(x_f_lr), label='polinominal 3')
plt.plot(x_f_lr, model4.predict(x_f_lr), label='Wielomian stopnia 4')
# plt.plot(x_f_lr, model5.predict(x_f_lr), label='Wielomian stopnia 5')
plt.plot(x_f_lr, model6.predict(x_f_lr), label='Wielomian stopnia 6')
# plt.plot(x_f_lr, model7.predict(x_f_lr), label='polinominal 7')
plt.plot(x_f_lr, model8.predict(x_f_lr), label='Wielomian stopnia  8')
# plt.plot(x_f_lr, model9.predict(x_f_lr), label='polinominal 9')
# plt.plot(x_f_lr, model10.predict(x_f_lr), label='polinominal 10')

plt.legend(fontsize=12)
plt.show()
```
