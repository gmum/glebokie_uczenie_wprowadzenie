```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.datasets import make_moons
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

Wizualizacje bariery decyzyjnej dla modelu logistycznej regresji.

```python
centers = [(-2, 0), (2, 0), (0, np.sqrt(20))]
X, y = make_blobs(n_samples=200, centers=centers, shuffle=False, random_state=3, cluster_std=0.9)

plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Dane zwierające trzy klasy", fontsize=15)
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
ovr_lr_clf = SVC(kernel='linear', C = 0.001, random_state=0)
ovr_lr_clf.fit(X,y)

plt.figure(figsize=(8, 5))
plot_decision_regions(X, y, ovr_lr_clf)
plt.title("Klasyfikacyjna metodą Logistycznej Regresji \n One versus Rest(ovr)", fontsize=15)
plt.show()
```

```python
from sklearn.multiclass import OneVsOneClassifier

ovo_lr_clf = OneVsOneClassifier(SVC(kernel='linear', C = 0.001, random_state=0))
ovo_lr_clf.fit(X,y)

plt.figure(figsize=(8, 5))
plot_decision_regions(X, y, ovo_lr_clf)
plt.title("Klasyfikacyjna metodą Logistycznej Regresji \n One versus ONE(ovo)", fontsize=15)
plt.show()
```
