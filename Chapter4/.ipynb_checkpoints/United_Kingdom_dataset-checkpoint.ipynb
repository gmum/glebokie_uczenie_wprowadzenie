```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

W poniższym przykładzie rozważmy zbiór danych dotyczących zależności między zakupem wyrobów tytoniowych i alkoholu w różnych
regionach Wielkiej Brytanii. 

```python
data_str = '''Region Alcohol Tobacco
North 6.47 4.03
Yorkshire 6.13 3.76
Northeast 6.19 3.77
East_Midlands 4.89 3.34
West_Midlands 5.63 3.47
East_Anglia 4.52 2.92
Southeast 5.89 3.20
Southwest 4.79 2.71
Wales 5.27 3.53
Scotland 6.08 4.51
Northern_Ireland 4.02 4.56'''
```

```python
from io import StringIO
df = pd.read_csv(StringIO(data_str), sep=r'\s+')
df.head(11)
```

```python
X = df.Tobacco
y = df.Alcohol
```

Wizualizacja danych dotyczących zależności między zakupem wyrobów tytoniowych i alkoholu w różnych regionach Wielkiej Brytanii. Na czerwono zaznaczony został jeden element, który będziemy traktować jako element odstający (ang. outlier).

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(X[-1:] ,y[-1:], 'or', ms=15)
ax.plot(X ,y,'o', ms=8)
ax.set_xlabel('Tobacco')
ax.set_ylabel('Alcohol')
ax.set_title('Sales in Several UK Regions')
ax.set_xlim([2.5, 4.7])
ax.set_ylim([3.8, 6.7])
plt.show()
```

Wizualizacja danych opisujących zależności między zakupem wyrobów tytoniowych i alkoholu w różnych regionach Wielkiej Brytanii bez elementu odstającego. Naszym celem będzie przewidzenie wartości dla punktu $X_0=3$.

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(X[:-1] ,y[:-1],'o', ms=8)
ax.vlines(3., 0, 7, colors='k', linestyles='--')
ax.set_xlabel('Tobacco')
ax.set_ylabel('Alcohol')
ax.set_title('Sales in Several UK Regions')
ax.set_xlim([2.5, 4.7])
ax.set_ylim([3.8, 6.7])
plt.show()
```

Zielona linia pokazuje prosta wyznaczoną za pomocą metody najmniejszych kwadratów.

```python
import sklearn.linear_model as lm
# We create the model.
lr = lm.LinearRegression()
# We train the model on our training dataset.
lr.fit(np.vstack(X[:-1]) ,y[:-1]);
print(lr.coef_)
print(lr.intercept_)
```

```python
x_f_lr = np.linspace(1., 5., 200)
f_lr = lambda x: lr.coef_ * x +lr.intercept_
y_f_lr = f_lr(x_f_lr)

fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.set_xlabel('Tobacco')
ax.set_ylabel('Alcohol')
ax.set_title('Sales in Several UK Regions')
ax.set_xlim([2.5, 4.7])
ax.set_ylim([3.8, 6.7])
ax.plot(x_f_lr, y_f_lr, 'g');
ax.plot(X[:-1] ,y[:-1],'o', ms=8)
ax.vlines(3., 0, 7, colors='k', linestyles='--')
# ax.plot(3., f_lr(3.), 'or', ms=10);
plt.show()
```

Wartość odpowiadająca predykcji leży na naszej prostej i oznaczona jest na czerwono.

```python
x_f_lr = np.linspace(1., 5., 200)
f_lr = lambda x: lr.coef_ * x +lr.intercept_
y_f_lr = f_lr(x_f_lr)

fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.set_xlabel('Tobacco')
ax.set_ylabel('Alcohol')
ax.set_title('Sales in Several UK Regions')
ax.set_xlim([2.5, 4.7])
ax.set_ylim([3.8, 6.7])
ax.plot(x_f_lr, y_f_lr, 'g');
ax.plot(X[:-1] ,y[:-1],'o', ms=8)
ax.vlines(3., 0, 7, colors='k', linestyles='--')
ax.plot(3., f_lr(3.), 'or', ms=10);
plt.show()
```

# Spróbujmy bezpośrednio zminimalizować funkcję kosztu 

Zdafiniuj funkcję opisującą błąd kwadratowy:

$$
\mathrm{error\_MSE}(a,b;(x,y)) =  \sum_{i=1}^{n} ( y_i - (ax_i+b) )^2.
$$

```python
def error_MSE(arg, *args):
    X, y = args[0], args[1]    
    a, b = arg
    return np.sum((y-(a*X+b))**2)
```

Zdafiniuj funkcję opisującą błąd basolutny:

$$
\mathrm{error\_MAE}(a,b;(x,y)) =  \sum_{i=1}^{n} |y_i - (ax_i+b)|.
$$

```python
def error_MAE(arg, *args):
    X, y = args[0], args[1]
    a, b = arg
    return np.sum(np.abs(y-(a*X+b)))
```

```python
from scipy import optimize

x0 = np.asarray((1, 1))
min_value_MSE = optimize.fmin_cg(error_MSE, x0, args=(X[:-1], y[:-1])) 
print(min_value_MSE)

min_value_MAE = optimize.fmin_cg(error_MAE, x0, args = (X[:-1],y[:-1]))
print(min_value_MAE)
```

Jak widać otrzymujemy podobne rozwiązania.

```python
x_f_lr = np.linspace(1., 5., 200)
f_lr_MSE = lambda x: min_value_MSE[0] * x + min_value_MSE[1]
y_f_lr_MSE = f_lr_MSE(x_f_lr)

f_lr_MAE = lambda x: min_value_MAE[0] * x + min_value_MAE[1]
y_f_lr_MAE = f_lr_MAE(x_f_lr)

fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.set_xlabel('Tobacco')
ax.set_ylabel('Alcohol')
ax.set_title('Sales in Several UK Regions')
ax.set_xlim([2.5, 4.7])
ax.set_ylim([3.8, 6.7])
ax.plot(x_f_lr, y_f_lr_MSE, 'g', label='$\sum (\hat f(x) - y)^2$')
ax.plot(x_f_lr, y_f_lr_MAE, 'b', label='$\sum |\hat f(x) - y|$')
# ax.plot(x_f_lr, y_f_lr, 'g');
ax.plot(X[:-1] ,y[:-1],'o', ms=8)
plt.legend()
plt.show()
```

Natomiast gdy wykorzystamy dane posiadające element odstający (w naszym przypadku pełne dane) to można zauważyć, ze takie podejście jest bardziej odporne na elementy odstające. 

```python
from scipy import optimize

x0 = np.asarray((1, 1))
min_value_MSE = optimize.fmin_cg(error_MSE, x0, args=(X, y)) 
print(min_value_MSE)

min_value_MAE = optimize.fmin_cg(error_MAE, x0, args = (X,y))
print(min_value_MAE)
```

```python
x_f_lr = np.linspace(1., 5., 200)
f_lr_MSE = lambda x: min_value_MSE[0] * x + min_value_MSE[1]
y_f_lr_MSE = f_lr_MSE(x_f_lr)

f_lr_MAE = lambda x: min_value_MAE[0] * x + min_value_MAE[1]
y_f_lr_MAE = f_lr_MAE(x_f_lr)

fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.set_xlabel('Tobacco')
ax.set_ylabel('Alcohol')
ax.set_title('Sales in Several UK Regions')
ax.set_xlim([2.5, 4.7])
ax.set_ylim([3.8, 6.7])
ax.plot(x_f_lr, y_f_lr_MSE, 'g', label='$\sum (\hat f(x) - y)^2$')
ax.plot(x_f_lr, y_f_lr_MAE, 'b', label='$\sum |\hat f(x) - y|$')
# ax.plot(x_f_lr, y_f_lr, 'g');
ax.plot(X ,y,'o', ms=8)
plt.legend()
plt.show()
```
