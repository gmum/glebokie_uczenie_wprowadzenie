```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import  metrics
```

# Wczytanie danych
W poniższym przykładzie rozważmy zbiór danych dotyczących zależności między zakupem wyrobów tytoniowych i alkoholu w różnych
regionach Wielkiej Brytanii. 

```python
data_str = '''Region Alcohol Tobacco
North 6.47 4.03
Yorkshire 6.13 3.76
Northeast 6.19 3.77
East_Midlands 4.89 3.34
West_Midlands 5.63 3.47
East_Anglia 4.52 2.92
Southeast 5.89 3.20
Southwest 4.79 2.71
Wales 5.27 3.53
Scotland 6.08 4.51
Northern_Ireland 4.02 4.56'''
```

```python
from io import StringIO
df = pd.read_csv(StringIO(data_str), sep=r'\s+')
df.head(11)
```

```python
X = df.Tobacco
y = df.Alcohol
```

Wizualizacja danych dotyczących zależności między zakupem wyrobów tytoniowych i alkoholu w różnych regionach Wielkiej Brytanii. 

```python
fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(X ,y,'o', ms=8)
ax.set_xlabel('Tobacco')
ax.set_ylabel('Alcohol')
ax.set_title('Sales in Several UK Regions')
ax.set_xlim([2.5, 4.7])
ax.set_ylim([3.8, 6.7])
plt.show()
```

```python
X=np.vstack(X)
model1 = linear_model.LinearRegression()
model1.fit(X, y)

model2 = make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression())
model2.fit(X, y)

model3 = make_pipeline(PolynomialFeatures(3), linear_model.LinearRegression())
model3.fit(X, y)

model4 = make_pipeline(PolynomialFeatures(4), linear_model.LinearRegression())
model4.fit(X, y)

model5 = make_pipeline(PolynomialFeatures(5), linear_model.LinearRegression())
model5.fit(X, y)

model6 = make_pipeline(PolynomialFeatures(6), linear_model.LinearRegression())
model6.fit(X, y)
```

```python
print(model1.score(X,y))
```

```python
x_f_lr = np.vstack(np.linspace(1., 5., 200))

fig, ax = plt.subplots(1, 1, figsize=(10,5))
ax.plot(X ,y,'o', ms=8)
plt.plot(x_f_lr, model1.predict(x_f_lr), label='$ y = w_1x + w_0 $')
plt.plot(x_f_lr, model2.predict(x_f_lr), label='$ y = w_2x^2 + w_1x + w_0 $')
plt.plot(x_f_lr, model3.predict(x_f_lr), label='$ y = w_3x^3 + w_2x^2 + w_1x + w_0 $')
plt.plot(x_f_lr, model4.predict(x_f_lr), label='$ y = w_4x^4 + w_3x^3 + w_2x^2 + w_1x + w_0 $')
plt.plot(x_f_lr, model5.predict(x_f_lr), label='$ y = w_5x^5 + w_4x^4 + w_3x^3 + w_2x^2 + w_1x + w_0 $')
plt.plot(x_f_lr, model6.predict(x_f_lr), label='$ y = w_6x^5 + w_5x^5 + w_4x^4 + w_3x^3 + w_2x^2 + w_1x + w_0 $')
ax.set_xlabel('Tobacco')
ax.set_ylabel('Alcohol')
ax.set_title('Sales in Several UK Regions')
ax.set_xlim([2.5, 4.7])
ax.set_ylim([3.8, 8.2])
plt.legend()
plt.show()
```

```python
regressors = {}
regressors['Model liniowy'] = model1
regressors['Model wielomianowy rzedu 2'] = model2
regressors['Model wielomianowy rzedu 3'] = model3
regressors['Model wielomianowy rzedu 4'] = model4
regressors['Model wielomianowy rzedu 5'] = model5
regressors['Model wielomianowy rzedu 6'] = model6

df = pd.DataFrame(columns=['Model', 'R^2', 'Explained variance score', 'Median absolute error', 'Mean squared error', 'Mean absolute errors'])

for key in regressors:
    regr = regressors[key]    
    df = df.append({'Model':key,
                    'R^2': metrics.r2_score(y_true = y, y_pred = regr.predict(X)),
                    'Explained variance score': metrics.explained_variance_score(y_true = y, y_pred = regr.predict(X)), 
                    'Median absolute error': metrics.median_absolute_error(y_true = y, y_pred = regr.predict(X)), 
                    'Mean squared error': metrics.mean_squared_error(y_true = y, y_pred = regr.predict(X)),
                    'Mean absolute errors':metrics.mean_absolute_error(y_true = y, y_pred = regr.predict(X)), 
                   } , ignore_index=True) 
    
```
