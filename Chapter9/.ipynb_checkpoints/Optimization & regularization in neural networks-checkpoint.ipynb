```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision.datasets import MNIST
from torchvision.transforms import ToTensor
```

```python
import matplotlib.pyplot as plt
import numpy as np

train_dataset = MNIST(".", download=True, transform=ToTensor())
train_loader = DataLoader(train_dataset, batch_size=128)
test_dataset = MNIST(".", download=True, transform=ToTensor(), train=False)
test_loader = DataLoader(test_dataset, batch_size=128)

fig, axes = plt.subplots(2, 5, figsize=(12, 2))
axes = axes.ravel()

for idx in range(10):
    axes[idx].imshow(train_dataset[idx][0].view(28, 28), cmap="gray")
    axes[idx].axis("off")
    # axes[idx].tight_layout()
plt.tight_layout()
plt.show()
```

```python
network = nn.Sequential(
    nn.Linear(784, 128),
    nn.ReLU(),
    nn.Linear(128, 128),
    nn.ReLU(),
    nn.Linear(128, 128),
    nn.ReLU(),
    nn.Linear(128, 10)
)
network = network.to("cuda")
```

```python
batch_size = 128
learning_rate = 3e-4


loss_fn = nn.CrossEntropyLoss()

# TODO: SGD, Adam, 
optimizer = torch.optim.Adam(network.parameters(), lr=learning_rate)

for epoch in range(100):
    
    for x, y in train_loader:
        x = x.view(-1, 784)
        preds = network(x.to("cuda"))
        loss_val = loss_fn(preds, y.to("cuda"))
        loss_val.backward()
        optimizer.step()
        optimizer.zero_grad()
    
    num_correct = 0
    for x, y in test_loader:
        x = x.view(-1, 784).to("cuda")
        preds = network(x)
        num_correct += (preds.argmax(-1) == y.to("cuda")).sum().item()
    print(f"Accuracy {num_correct / len(test_dataset)}")
        
```
