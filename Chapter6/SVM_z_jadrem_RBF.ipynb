```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
```

Wizualizacja bariery decyzyjnej modelu SVM z jądrem RBF z parametrem C = 1 oraz różnymi parametrami gamma. Jak widzimy, dzięki zmianie parametru opisującego jądro gaussowskie możemy kontrolować poziom overfitingu.

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=.1, random_state=42)

plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Dwie klasy reprezentujące dwie nieliniowe grupy", fontsize=15)
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
```

# Linear SVM

```python
from sklearn.svm import SVC

C = 1
gamma = 0.01
svm_clf = SVC(kernel='rbf', random_state=0, gamma=gamma, C=C)
svm_clf.fit(X,y)

plt.figure(figsize=(8, 5))
plot_decision_regions(X, y, svm_clf)
plt.title("Klasyfikacyjna metodą liniowego SVM z parametrm C = " + str(C) + " gamma = " + str(gamma), fontsize=15)
plt.show()
```
